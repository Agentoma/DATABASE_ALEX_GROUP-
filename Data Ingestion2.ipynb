```python
# Important the libraries
import requests, json
import pandas as pd
import datetime as dt
```

```python
# state the resource ID
resource_id = '1666aacb-0503-4e8c-9acf-156a0e199ee0'
Nrecords =  100 # number of data sets to be returned
api_url = 'https://data.ontario.ca/en/api/3/action/datastore_search?resource_id={}&limit={}'.format(resource_id, Nrecords) # the URL to be used


```

```python
# Headers to be used in the request to the API 
headers = {'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 14_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.5 Safari/605.1.15'}
```

```python
response = requests.get(api_url, headers=headers) # send the request to the API
```

```python
print(response.status_code) # print the status code of the response
```

```python
data = response.json() # convert the response to a JSON object
data
```
